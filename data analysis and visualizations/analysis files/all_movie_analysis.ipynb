```python
# import dependencies
import pandas as pd
import os
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the path and import the data
all_movie_data = os.path.join('../all_movie_info.csv')
movie_df = pd.read_csv(all_movie_data)
```

```python
# View data
movie_df.head()
```

```python
# Drop unnecessary columns
clean_movie_df = movie_df.drop(columns=['id','imdb_id'])
clean_movie_df
```

```python
# Check for null values
clean_movie_df.isnull().sum()
```

```python
# Drop rows with null values and verify removal
clean_movie_df.dropna(axis=0,how="any",inplace=True)
clean_movie_df.isnull().sum()
```

```python
# View data
clean_movie_df
```

```python
# Create dataframe to hold the winners
winner_df = clean_movie_df.loc[(clean_movie_df['winner']==True)]
winner_df.head()
```

```python
# Count the number of winners
num_winners = winner_df['popularity'].count()
num_winners
```

```python
# Explore popularity level column
clean_movie_df['popularity_level'].unique()
```

```python
# Make barchart displaying the oscar winners by popularity level 
```

```python
# Create dataframe to hold winners with a high popularity level
winner_high_pop = winner_df.loc[(winner_df['popularity_level'] == 'High')]
winner_high_pop.head()
```

```python
# Find the number of winners for high popularity
num_high_winners = winner_high_pop['winner'].count()
num_high_winners
```

```python
# Create Dataframe to hold winners with moderately high popularity level
winner_mod_high_pop = winner_df.loc[(winner_df['popularity_level'] == 'Moderately High')]
winner_mod_high_pop.head()
```

```python
# Find the number of winners for moderately high popularity
num_mod_high_winners = winner_mod_high_pop['winner'].count()
num_mod_high_winners
```

```python
# Create Dataframe to hold winners with medium popularity level
winner_med_pop = winner_df.loc[(winner_df['popularity_level'] == 'Medium')]
winner_med_pop.head()
```

```python
# Find the number of winners for medium popularity
num_med_winners = winner_med_pop['winner'].count()
num_med_winners
```

```python
# Create Dataframe to hold winners with low popularity level
winner_low_pop = winner_df.loc[(winner_df['popularity_level'] == 'Low')]
winner_low_pop.head()
```

```python
# Find the number of winners for low popularity
num_low_winners = winner_low_pop['winner'].count()
num_low_winners
```

```python
# Create a barchart to show the oscar winners by popularity level

# creating the dataset
data = {'High': num_high_winners, 'Moderately High': num_mod_high_winners, 'Medium ': num_med_winners,
        'Low': num_low_winners}

courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='darkolivegreen',
        width = 0.4)
 
plt.xlabel("Popularity Level ")
plt.ylabel("Number of Oscar Winners")
plt.title("Number of Oscar Winners by Popularity Level")
plt.show()


fig.savefig('../all movie analysis/winner_by_pop.png')  # save the figure to file
plt.close(fig)    # close the figure window
```
