```python
# Find average of the high popularity vote average
mean_high_pop = high_pop['vote_average'].mean()
mean_high_pop
```

```python
# Find average of the moderately high popularity vote average
mean_mod_high_pop = mod_high_pop['vote_average'].mean()
mean_mod_high_pop
```

```python
# Find average of the medium popularity vote average
mean_medium_pop = medium_pop['vote_average'].mean()
mean_medium_pop
```

```python
# Find average of the low popularity vote average
mean_low_pop = low_pop['vote_average'].mean()
mean_low_pop
```

```python
# Create a barchart to show the average vote by popularity

# creating the dataset
data = {'High': mean_high_pop, 'Moderately High': mean_mod_high_pop, 'Medium ': mean_medium_pop,
        'Low':mean_low_pop}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='cornflowerblue',
        width = 0.4)
 
plt.xlabel("Popularity ")
plt.ylabel("Vote Average")
plt.title("Average Votes by Popularity Level")
plt.show()

#plt.savefig('avg_votes_by_pop.png',bbox_inches='tight')
#plt.close(figure_object)

fig.savefig('../Margo_group_work/vot_avg.png')   # save the figure to file
plt.close(fig)    # close the figure window
```
