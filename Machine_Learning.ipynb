```python
# Import dependecies
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# Import data for pre-processing
df = pd.read_csv('all_movie_info.csv')
df.head(10)
```

```python
# Drop irrelevant or redundant columns
df = df.drop(['id', 'imdb_id', 'popularity', 'budget', 'revenue', 'homepage', 'tagline',
              'overview', 'index', 'year_film', 'year_ceremony', 'ceremony', 'name', 'film', 'keywords',
              'release_year', 'original_title', 'category'], axis = 1)
df.head(10)
```

```python
# Extract movie release months from the 'release_date' values.
release_month = []
for release_date in df['release_date']:
    my_str = release_date
    my_list = my_str.split('/')
    my_int_list = [int(i) for i in my_list]
    my_month = my_int_list[0]
    release_month.append(my_month)
    
release_month[0:10]
```

```python
df['release_month'] = release_month
df = df.drop(['release_date'], axis = 1)
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['popularity_encoded'] = encoder.fit_transform(df['popularity_level'])
df.head(10)

# Note that encoder.inverse_transform(df['popularity_encoded']) will return the original
# string values
```

```python
df = df.drop(['popularity_level'], axis = 1)
df.head(10)
```

```python
df['cast'] = df['cast'].str.split("|")
df['cast'][0]
```

```python
df['director'] = df['director'].str.split("|")
df['director'][0]
```

```python
df['genres'] = df['genres'].str.split("|")
df['genres'][0]
```

```python
df['production_companies'] = df['production_companies'].str.split("|")
df['production_companies'][0]
```

```python
df.shape
```

```python
# Because some of the column values are lists of strings, rather than string or numeric types, 
# we separate the column and create a stacked Pandas Series object. This is then converted to a 
# dataframe.
new_df = df.drop('cast', axis = 1).join(
             pd.get_dummies(
             pd.DataFrame(df.cast.tolist(), df.index).stack()
             ).astype(int).groupby(level=0).sum()
        )
new_df.head(10)
```

```python
new_df.shape
```

```python
new_df = new_df.drop('genres', axis = 1).join(
             pd.get_dummies(
             pd.DataFrame(df.genres.tolist(), df.index).stack()
             ).astype(int).groupby(level=0).sum()
        )
new_df.head(10)
```

```python
new_df.shape
```

```python
# new_df = new_df.drop('director', axis = 1).join(
#              pd.get_dummies(
#              pd.DataFrame(df.director.tolist(), df.index).stack()
#              ).astype(int).groupby(level=0).sum()
#         )
director_dummies = pd.get_dummies(new_df['director'].apply(pd.Series).stack()).groupby(level=0).sum()
director_dummies.columns = [f"{col}_director" for col in director_dummies.columns]

# concatenate the one-hot encoded columns back into the original dataframe
new_df = pd.concat([new_df.drop(['director'], axis=1), director_dummies], axis=1)
new_df.head(10)
```

```python
new_df.shape
```

```python
new_df = new_df.drop('production_companies', axis = 1).join(
             pd.get_dummies(
             pd.DataFrame(df.production_companies.tolist(), df.index).stack()
             ).astype(int).groupby(level=0).sum()
        )
new_df.head(10)
```

```python
new_df.shape
```

```python
new_df['winner']
```

```python
# Convert null values to False
new_df['winner'] = new_df['winner'].fillna(0).astype(int)
```

```python
# Convert boolean values to binary values
new_df['winner'] = new_df['winner'].astype(int)
```

```python
new_df['winner']
```

```python
y = new_df['winner']
X = new_df.copy()
X.head(10)
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
X = X.drop(columns = 'winner')
X.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
import tensorflow as tf

# Create the architecture for the neural network
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=500, activation="relu", input_dim = 5416))

# Second hidden layer
# nn.add(tf.keras.layers.Dense(units=5, activation = "relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation = "sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Results
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)

### The following code is from ChatGPT to attempt to correct the type error when evaluating the model's performance.
# for i in range(len(X_test_scaled)):
#     # Get the input sample from the test set
#     input_sample = X_test_scaled[i]
 
#     # Use the model to make a prediction on the input sample
#     prediction = nn.predict(input_sample.reshape(1, -1))[0][0]
 
#     # Round the prediction to the nearest integer (0 or 1)
#     binary_prediction = int(round(prediction))
 
#     # Display the input sample and the binary prediction
#     print(f"Input sample: {input_sample}\nBinary prediction: {binary_prediction}\n")
```

```python
from sklearn.metrics import confusion_matrix

y_pred = nn.predict(X_test_scaled)
confusion = confusion_matrix(y_test, y_pred)
confusion

# [[true negative  false positive]
#  [false negative true positive]]
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# Calculate precision and recall
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

print(f"Loss: {model_loss}, Accuracy: {model_accuracy}, Precision: {precision}, Recall: {recall}")
```

```python
# Now that we have the results from the neural network, train and evaluate a random forest model for comparision.
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.linear_model import LogisticRegression

balanced_rf_classifier = BalancedRandomForestClassifier(random_state = 1)
balanced_rf_classifier.fit(X_train, y_train)

# Fit the Logistic Regression model to the resampled training data
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_train, y_train)
```

```python
# Evaluate the random forest model's performance
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```
